```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([1, 8])
y = np.array([3, 10])
plt.plot(x, y)
plt.show() # removes type info message above the plot
```

```python
plt.plot(x, y, 'o-')
```

```python
x = np.array([1, 2, 5, 8])
y = np.array([3, 8, 2, 10])
plt.plot(x, y, '*')
```

```python
plt.plot(x, '--r')
```

```python
plt.plot(x, y, 'o:', ms=8, linewidth=3)
```

```python
y1 = np.array([3, 8, 2, 10])
y2 = np.array([6, 2, 8, 11])
plt.plot(y1)
plt.plot(y2)
plt.show()
```

```python
x = np.linspace(-5, 5, 200)
plt.plot(x, np.sin(x), label='sin')
plt.plot(x, np.cos(x), label='cos')
plt.xlim([-2, 5])
font1 = {'family': 'serif', 'color': 'blue', 'size': 18}
font2 = {'family': 'serif', 'color': 'red', 'size': 13}
plt.xlabel('x axis', fontdict=font2)
plt.ylabel('y axis', fontdict=font2)
plt.title('Sample Graph', fontdict=font1, loc='right')
plt.grid(axis='y') # plt.grid()
plt.legend()
plt.show()
```

```python
# plot 1
x = np.array([0, 1, 2, 3])
y = np.array([3, 9, 5, 10])
plt.subplot(1, 2, 1)
plt.plot(x, y)

# plot 2
x = np.linspace(0, 10, 100)
plt.subplot(1, 2, 2)
plt.plot(x, np.sin(x))
plt.show()
```

```python
# plot 1
x = np.array([0, 1, 2, 3])
y = np.array([3, 9, 5, 10])
plt.subplot(2, 1, 1)
plt.plot(x, y)

# plot 2
x = np.linspace(0, 10, 100)
plt.subplot(2, 1, 2)
plt.plot(x, np.sin(x))

plt.suptitle('Sample Graphs')
plt.show()
```

```python
# scater graphs

x = np.random.normal(170, 10, 200)
y = np.random.normal(100, 10, 200)
plt.scatter(x, y)
plt.show()
```

```python
# bar graph

x = np.array(['A', 'B', 'C', 'D'])
y = np.array([2, 5, 1, 10])
plt.barh(x, y, color='#4ACF40') # barh - horizontal type of bar graph
```

```python
x = np.random.normal(150, 10, 200)
plt.hist(x)
```

```python
y = np.array([30, 40, 20, 10])
mylabels = ['Food', 'Tools', 'Furniture', 'Household goods']
plt.pie(y, labels=mylabels, startangle=90, explode=[0.2, 0, 0, 0], shadow=True, autopct='%1.1f%%')
plt.show()
```

```python
from numpy.random import exponential
x = exponential(5, 20)
print(x)
plt.plot(x)
plt.show()
```

```python
from matplotlib import axes
fig, axes = plt.subplots()
axes.plot(exponential(5, 20))
axes.set_title('Sample graph')
```

```python
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(10, 5))
for row, row_axes in enumerate(axes):
  for column, ax in enumerate(row_axes):
    ax.plot(exponential(column + 1, 20))
    ax.set_title('Column {} row {}'.format(column + 1, row + 1))
  fig.tight_layout()
```
